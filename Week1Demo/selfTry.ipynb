```python
hist = {}
for token in parsed_txt:
    if not token.orth_ in hist.keys():
        hist[token.orth_] = 1
    else:
        hist[token.orth_] += 1
```

```python
from operator import itemgetter

sorted_item = sorted(hist.items(),
                    reverse = True,
                    key = itemgetter(1))
for i, j in sorted_item:
    print(i, ': ', j)
```

```python
from collections import Counter
hist = Counter()
for token in parsed_txt:
    word = token.orth_
    hist[word] += 1
    
sortWord = sorted(hist.items(),
                 reverse = True,
                 key = itemgetter(1))

for i,j in sortWord:
    print(i, ':', j)
```

```python

hist = Counter()
for token in parsed_txt:
    if not token.is_stop or token.pos_ == 'PUNCT' or token.pos_ == 'SPACE':
        hist[token.orth_] += 1
        
sort_word = sorted(hist.items(),
                  reverse = True,
                  key = itemgetter(1))

for i,j in sort_word:
    print(i, ': ', j)
```
